```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('loan_data.csv') #https://www.kaggle.com/datasets/deependraverma13/lending-club-loan-data-analysis-deep-learning?resource=download
df.sample(10)

print("shape: ", df.shape)
df.describe().T.round(2)
```

```python
df.isnull().sum()
```

```python
num_cols= df.select_dtypes(include='number').columns
cat_cols= df.select_dtypes(exclude='number').columns
target_cols='not.fully.paid'

print("Numerical Column: ", num_cols)
print("Catrgorical Column: ", cat_cols)
print("Target Column: ", target_cols)
```

```python
#for one hot encoding

df = pd.get_dummies(df, columns=['purpose'], drop_first=True)
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = df.drop(columns=[target_cols])
y = df[target_cols]

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Question 02: [ Marks 20 ]

## Design a neural network using nn.Module.

## Justify:

## Write Answer 02:


```python
import torch
import torch.nn as nn

class LoanModel(nn.Module):
    def __init__(self, input_size):
        super(LoanModel, self).__init__()
        self.fc1 = nn.Linear(input_size, 16)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(16, 8)
        self.fc3 = nn.Linear(8, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        return x
```

## Implement a full training loop:

## Write Answer 04:

```python
import torch

# Features
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

# Labels (important: reshape)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)
```

```python
epochs = 100

for epoch in range(epochs):
    model.train()

    # Forward pass
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()

    # Update weights
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```

#In PyTorch, gradients do not reset automatically—they keep adding up.
#If we don’t clear them, the model will update weights using wrong values.

#If the learning rate is too high:

#i) The model takes very big steps
#ii) It may skip the correct solution
#iii) Loss may go up and down instead of decreasing

#If the learning rate is too small:

The model learns very slowly.
It takes too many epochs.
Training becomes inefficient

#We define layers in __init__ because:

They are created once.
Their weights are updated during training

#If we define them in forward():

New layers will be created every time.
