##Manual Process

```python
x=torch.tensor(2.0)
y=torch.tensor(0.0)

w=torch.tensor(1.0)
b=torch.tensor(0.0)
```

```python
#BCE

def binary_cross_entropy(prediction, target):
  epsilon=1e-8
  prediction=torch.clamp(prediction, epsilon, 1-epsilon)
  return -(target*torch.log(prediction)+(1-target)*torch.log(1-prediction))
```

```python
#forward pass

z=w*x+b
y_pred = torch.sigmoid(z)

#computer BCE

loss=binary_cross_entropy(y_pred,y)
```

```python
loss
```

```python
#derivative
#1

dloss_by_pred =  (y_pred-y)/(y_pred*(1-y_pred))

#2
dy_pred_dz=y_pred*(1-y_pred)
dz_dw=1
dz_db=0

dL_dw = dloss_by_pred * dy_pred_dz*dz_dw
dL_db = dloss_by_pred * dy_pred_dz*dz_db
```

```python
dL_db
```

```python
dL_dw
```

##Using Autograd

```python
x=torch.tensor(2.0)
y=torch.tensor(0.0)

w=torch.tensor(1.0, requires_grad=True)
b=torch.tensor(0.0, requires_grad=True)
```

```python
z=w*x+b
```

```python
z
```

```python
y_pred=torch.sigmoid(z)
y_pred
```

```python
loss=binary_cross_entropy(y_pred,y)
loss
```

```python
loss.backward()
```

```python
w.grad
```

```python
b.grad
```
